```python
Texas_covid_clean=pd.read_csv("Texas_covid_clean.csv")
tx_visualization=pd.read_excel("tex visualization.xlsx")
tx_visualization["year"]=pd.to_datetime(tx_visualization["year"])
Texas_covid_clean["Date"]=pd.to_datetime(Texas_covid_clean["Date"],yearfirst=True)
fig, ax1 = plt.subplots()
plt.ylim(0, 140000)
ax1.plot(tx_visualization["year"],tx_visualization["flight number "],color="green",label='Flight number')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Number of flight')
ax1.legend(loc=2)
ax2 = ax1.twinx()
ax2.plot(Texas_covid_clean["Date"], Texas_covid_clean["New Cases"],label='New covid cases')
ax2.set_ylabel('new covid cases')
ax2.legend(loc=9)

X = [18350, 18350]
Y = [0, 140000]
plt.plot(X,Y,"--")
plt.ylim(0, 30000)
plt.title('number of flight VS Covid case in Texas')
plt.show()
```
