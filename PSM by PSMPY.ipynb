```python
import pandas as pd
```

```python
df_ny = pd.read_csv('NY_Flight_Covid.csv')
```

```python
df_tx = pd.read_csv('TX_Flight_Covid.csv')
```

```python
df_ny.shape
df_ny.head()
```

```python
df_tx.head()
```

```python
df_ny['NY(1)_TX(0)'] = 1
df_tx['NY(1)_TX(0)'] = 0
frame = [df_ny, df_tx]
df = pd.concat(frame, ignore_index=True)

```

```python
# df['DEP_DELAY_NEW'] = df['DEP_TIME'] - df['CRS_DEP_TIME']
# df['ARR_DELAY_NEW'] = df['ARR_TIME'] - df['CRS_ARR_TIME']

#create punctuality label, I assume it is punctual departure if the absolute value of diff is less than 10 min
df['punctuality'] = 0
df.loc[abs(df.DEP_DELAY_NEW) <= 10, 'punctuality'] = 1

```

```python
df.tail(50)
```

```python
df['index'] = df.index
```

```python
from psmpy import PsmPy
from psmpy.functions import cohenD
from psmpy.plotting import *
```

```python
df = df[['YEAR','MONTH','DAY_OF_WEEK','ARR_DELAY_NEW','CANCELLED','NY(1)_TX(0)','index','DEP_DELAY_NEW','CARRIER_DELAY','SECURITY_DELAY','LATE_AIRCRAFT_DELAY','punctuality']]
df = df.dropna()
len(df)
```

```python
df = df.sample(frac=1, random_state = 265)
df = df.head(20000)
```

```python
psm = PsmPy(df,treatment = 'NY(1)_TX(0)', indx='index', exclude = ['DAY_OF_WEEK','CANCELLED','YEAR','MONTH','punctuality'])
```

```python
psm.logistic_ps(balance = True)
```

```python
psm.predicted_data
```

```python
psm.knn_matched(matcher='propensity_logit', replacement=False, caliper=None)
```

```python
psm.plot_match(Title='Side by side matched controls', Ylabel='Number of airlines', Xlabel= 'Propensity logit', names = ['treatment', 'control'], save=True)
```

```python
psm.matched_ids
```

```python
psm.effect_size_plot(save=True)
```
