```python
#a function to count survival in each class
def survival(dataf):
    count = 0
    for ele in dataf['Survived']:
        if ele == 1:
            count += 1
    return count
```

```python
#Q5 
#Variance and Std:
def var(datalist):
    meanval = np.mean(datalist)
    n = len(datalist)
    sumval = 0
    for ele in datalist:
        sumval = (ele - meanval)**2 + sumval
        
    return sumval/(n-1)     
```

```python
#Covariance for correlation
def covar(x,y):
    meanx = np.mean(x)
    meany = np.mean(y)
    sumval = 0
    n = len(x)
    for i in range(len(x)):
        sumval = (x[i]-meanx)*(y[i]-meanx) + sumval
    cov = sumval/(n-1)
    return cov
```

```python
def correl(x,y):
    cor = covar(x,y)/(np.sqrt(var(x))*np.sqrt(var(y)))
    return cor
```

```python
#Make a matrix in a order of Survived, Age, SibSp, Parch, Fare, NotAlone
def matrix(numlist):
    matrix = []
    for i in numlist:
        lis = []
        for j in numlist:
            ele = correl(titanic_df[i],titanic_df[j])
            lis.append(ele)
        matrix.append(lis)
    
    return matrix


            
```

```python
#input df
numdf = titanic_df[['Survived','Age','SibSp','Parch','Fare','NotAlone']]
```

```python
#correlation matrix (round 3 decimal places)
cor_matrix = np.round(matrix(numdf),3)
```

```python
#Q9

#1.
namelis = list(titanic_df['Name'])
titlelis = []
for ele in namelis:
    ele = ele.split(' ')
    for i in ele:
        if '.' in i:
            titlelis.append(i)
    
titlelis = set(titlelis)
```

```python
#2.

surv_rate_lis = []
surv_count_lis = []
for t in titlelis:  
    #create a dataframe that contains the specific title
    newdf = titanic_df[titanic_df['Name'].str.contains(t)]
    #dataframe for survivals
    newdf1 = newdf[newdf['Survived'] == 1]
    surv_count = len(newdf1)
    surv_rate = surv_count/len(newdf)
    surv_rate_lis.append(surv_rate)
    surv_count_lis.append(surv_count)
        
    
```

```python
#I create a dataframe for viewing different title's survived rate
surv_df = pd.DataFrame(surv_rate_lis, columns=['Survived Rate'])
surv_df['Title'] = list(titlelis)
surv_df['Survived Amount'] = surv_count_lis
surv_df = surv_df.sort_values(by = ['Survived Amount'], ascending=False)
surv_df
```

Observation: All 'Ms.','Lady.','Mme.','Mlle.','Countess.' survived. All 'Jonkheer.', 'Rev.','Don.','Capt.' died.
Mrs has the higher survival rate than Mr. Miss and Mrs's survival rates are high. Col and Mr's survival rates are low.
