### Q1

The  Correlates of War (COW) project is a study about the international relationships and especially focuses on the history of wars. The idea behind of the project is to use accurate and reliable datas in scientific studies thus it follows the principles innclude transparency of data, accurate collections and well documentations. The history of the COW project traces back to the year of 1963, found by J.David Singer, from the University of Michigan and thus has its nature as an academic resources and strictly not for any commercial uses. 

The dataset are published by the directors as well as in cooperation with advisory board with both being responsible for keeping track of the data sets. There were in total of 5 past directors who were involved and currently the project is having co-directors: Jeff Carter from Appalachian State University and Scott Wolford from the University of Texas. Along with other people such as the associate director, D. Scott Bennett from Penn-State  University, 9 members in the advisory board, and many data hosts including people from Texas A&M,University of  Wisconsin, UC-Davis and many more. There are many data sets with unique features that makes up the project. This  includes "State System Membership" that marks the change of composition over time; "COW War Data" that serves as  the datalist that contains a list of wars since 1816; The "Militarized Interstate Disputes" dataset that keeps on track when one state threatens another; and most importantly, the "National Material Capabilities" which is used  in this report that contains important annual values espeically the six indicatos;and many more datasets such as  religion data, alliances, territory changes and so on.

### Q2

The "v. 5.0 of National Material Capabilities dataset" is currently not the latest version as there is another v.6 version avaliable. As of the v.5 version, the main updates comparing to the v.4 version is the update at the year  of 2017, by extending the data for the six components(NMC) to 2012. In addition, urban population data was changed  significantly as now is the data provided by the UN. Last but not least, the procedure in calculation of the  Composite Index of National Capabilities, also known as CINC, which is one of the variable in the data sheet, is  also added. All of these information can be accessed in the "2017" updated sub-section. Speaking of the variables in the data sheet, here is a brief explanation. 

Starting with the “stateabb” code which is a 3 letter abbreviation for country names, such as "USA", "CAN"; each of the countries also has it's unique COW country code and labeled as “ccode”; The variable “year” stands for the year of observation of this specific country related to the study; it then comes to the immportant measures section with “irst” being the iron and steel production and unit is "thousands tons"; “milex” is the measure of military  expenditures and the unit here will be tricky, thousands of Pounds for  1816-1913 and US Dollars from 1914 and on; “milper” is the military personnel number being recorded in unit of thousands; “pec” simply being the energy consumption with its unit in thousands of coal-ton and equivalent amounts; “tpop” stands for the total population where as “upop” is the urban population, where people living in cities with 100,000 population or more, both has an unit of thousands; lastly it is the “cinc”, the new implented variable starting from version 5, which is the  composite index of national aapability score with an unique calculation method usually being a decimal number that is less than 1.

The supplementary data file for v.5 contains all core variables included in the v.5 file of NMC datset while having additional information, especially the source of data and its quality codes. Supplementary information is added for the six main variables including "milex", "milper", "irst", "pec", "tpop" and "upop". While "milex" and "milper"  both have a source and note file, that are simply named with an addition of "source", or "note" after the variable name; "irst", "pec", and "tpop" are supplemented by another set of code information, quality code and the anomaly code which are also named by addition of strings; lastly, "upop" has all the additional materials listed above as  well as a "growth rate" variable with its source. Generally looking at the data, most sources are coming from recognnizeble sources such as "U.S. Department of Commerce and Bureau of the Census", "Iron and Stell in the 19th  Century America", "Historical Statistics of the United States", or "UN Energy Statistics Database". However, one  problem that may cause researchers to spend more time on is the consistency of the sources that are used. Because even for the same country, the sources varies a lot as the time goes and the accuracy may not be as high if one single source is used for the entire time. Also there are many data sources are left blank, indicating a missing information, while this might not be a big problem for discriptive data such as sources, notes, when it comes to the numerical data such as the growth rate of urban population, it potentially causes more trouble for scientists to analyze the causal relationships and making future predictions.

"pec" also has the positive correlation with both "tpop" and "upop" because based on theory, the more poeple there are, especially for those in the urbanized area, more energy will be consumed. Last but not least, "milper" might have a correlation with "tpop" because more people the avaliable, the more armed force there could be. For exmaple the "irst" and the "upop" because naturally, urbanized places uses more ennergy and when it includes the factor of population, the connection won't be as strong as it's being displayed.
```python
#Standardize
from sklearn.preprocessing import MinMaxScaler

X = cow_df_2012[['milex','milper','irst','pec','tpop','upop','cinc']]
scaler = MinMaxScaler()
scaler.fit(X)
X = scaler.transform(X)
cow_df_2012[['milex','milper','irst','pec','tpop','upop','cinc']] = X
```

```python
#Make a matrix 
from scipy.spatial import distance
#manhattan
def man_matrix(numlist):
    matrix = []
    for i in numlist:
        lis = []
        for j in numlist:
            ele = distance.cityblock(i,j)
            lis.append(ele)
        matrix.append(lis)
    return matrix

#euclidean
def euc_matrix(numlist):
    matrix = []
    for i in numlist:
        lis = []
        for j in numlist:
            ele = distance.euclidean(i,j)
            lis.append(ele)

        matrix.append(lis)
    return matrix

#country_name tag to track pairs
def name_matrix(countrylist):
    matrix = []
    for i in countrylist:
        lis = []
        for j in countrylist:
            ele = i +' '+ j
            lis.append(ele)
        matrix.append(lis)
    return matrix
```

```python
#pair countries in the order of calculating euclidean distance
countrylist = name_matrix(cow_df_2012['stateabb'])

#195*195 euclidean and manhattan distance
euc = euc_matrix(X)
manh = man_matrix(X)

#dataframe of euclidean and manhattan distance
data = {'Euclidean': euc_matrix(X),'manhattan':man_matrix(X)}
df = pd.DataFrame(data=data)
```

```python
#the dictionary contains country pair as key and distance as value
#Ex: {'USA USA': 0.0}
#euclidean:
eudict = {}
for i in range(195):
    for j in range(195):
        eudict[countrylist[i][j]] = euc[i][j]

eulist = sorted(eudict.items(), key=lambda item: item[1])
top10_euc = eulist[195:215]
top10_euc = [x[0] for x in top10_euc] #get rid of the dictionary values
top10_euc = top10_euc[::2]
 
#manhattan:
manhdic = {}
for i in range(195):
    for j in range(195):
        manhdic[countrylist[i][j]] = manh[i][j]

manhlist = sorted(manhdic.items(), key=lambda item: item[1])
top10_manh = manhlist[195:215]
top10_manh = [x[0] for x in top10_manh]
top10_manh = top10_manh[::2]

print('top10 Euclidean distance pairs are:',top10_euc)
print('top10 Manhattan distance pairs are:',top10_manh)
```
